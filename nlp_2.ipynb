```python
file_path = '/content/drive/MyDrive/data/wiki_us.txt'
with open(file_path, 'r') as file:
    text = file.read()

print(text)
```

Creating a Doc Container

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp(text)
print(doc)
```

```python
print(len(doc))
print(len(text))
```

```python
for token in text[:10]:
  print(token)
```

```python
for token in doc[:10]:
  print(token)
```

```python
for token in text.split()[:10]:
  print(token)
```

```python
i = 5
for token in doc[i:8]:
  print(f"SpaCy Token {i}: {token}")
  i=i+1
```
